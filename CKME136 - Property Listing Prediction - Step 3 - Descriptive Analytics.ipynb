### 3a. Imports for Descriptive Analytics

```python
# imports for visualization
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

%matplotlib inline
pd.set_option('display.max_rows', 500)
```

```python
# read in descriptive analytics dataset
df_all_features_desc_dataset = pd.read_csv('cleaned_combined_listing_descriptive_analytics_v1.csv', index_col=0)
```

```python
# see a sample of the dataset
df_all_features_desc_dataset.head()
```

```python
# describe dataset
pd.options.display.float_format = '{:.2f}'.format
df_all_features_desc_dataset.describe()
```

### 3b. Histogram of target variable

```python
# histogram on dependent variable
figure(figsize=(20, 10))
plt.hist(df_all_features_desc_dataset['listing_price_int'], bins = 1000)
plt.title('Listing Price Histogram')
plt.xlabel('Listing Price')
plt.ylabel('Property Count')
ax = plt.gca()
ax.spines['left'].set_position('zero')
ax.get_xaxis().get_major_formatter().set_scientific(False)
plt.draw()
```

```python
# investigate extreme outlier
df_all_features_desc_dataset.where(df_all_features_desc_dataset['listing_price_int'] > 20000000).dropna()
```

```python
#filtered histogram to omit outlier beyond comparable sold prices
df_filtered_listing_price = df_all_features_desc_dataset[['listing_price_int']].where(df_all_features_desc_dataset['listing_price_int'] < 13000000).dropna()

#plot
figure(figsize=(20, 10))
plt.hist(df_filtered_listing_price['listing_price_int'], bins = 1000)
plt.title('Filtered Listing Price Histogram')
plt.xlabel('Listing Price')
plt.ylabel('Property Count')
ax = plt.gca()
ax.spines['left'].set_position('zero')
ax.get_xaxis().get_major_formatter().set_scientific(False)
plt.draw()
```

```python
# descriptive analytics of list prices
df_filtered_listing_price.describe()
```

```python
# view attributes of low skew listings
df_filtered_low_prices = df_all_features_desc_dataset.where(df_all_features_desc_dataset['listing_price_int'] < 200000).dropna()
df_filtered_low_prices[['listing_url', 'listing_price_int', 'comparable_sold_price_1_int', 'postal_code_fsa', 'community', 'building_type']].head(20)
```

```python
# count of low skew by building types
df_filtered_low_prices_by_building_type = df_filtered_low_prices.groupby('building_type').count().reset_index().iloc[:,0:2]

figure(figsize=(10, 5))
plt.barh(df_filtered_low_prices_by_building_type['building_type'],df_filtered_low_prices_by_building_type['listing_url'])
plt.title('Listings Less than $200K by Building Type')
plt.xlabel('Property Count')
plt.ylabel('Building Type')
plt.draw()
```

### 3c. Price distributions by dwelling type

```python
# Price distribution of all listings
df_clean_prices = df_all_features_desc_dataset[((df_all_features_desc_dataset['listing_price_int'] > 200000) & (df_all_features_desc_dataset['listing_price_int'] < 13000000)) | (df_all_features_desc_dataset['building_type'] == 'Apartment')].dropna()
df_clean_prices.iloc[:,:4].describe()
```

```python
# Price distribution of apartment listings
df_clean_prices = df_all_features_desc_dataset[((df_all_features_desc_dataset['listing_price_int'] > 200000) & (df_all_features_desc_dataset['listing_price_int'] < 13000000) & (df_all_features_desc_dataset['building_type'] == 'Apartment'))].dropna()
df_clean_prices.iloc[:,:4].describe()
```

```python
# Price distribution of apartment listings
df_clean_prices = df_all_features_desc_dataset[((df_all_features_desc_dataset['listing_price_int'] > 200000) & (df_all_features_desc_dataset['listing_price_int'] < 13000000) & (df_all_features_desc_dataset['building_type'] != 'Apartment'))].dropna()
df_clean_prices.iloc[:,:4].describe()
```

```python
df_building_type = df_all_features_desc_dataset.groupby('building_type').count().sort_values('listing_url', ascending = False).reset_index().iloc[:,0:2]
figure(figsize=(20, 10))
plt.barh(df_building_type['building_type'],df_building_type['listing_url'])
plt.title('Listings by Building Type')
plt.xlabel('Property Count')
plt.ylabel('Building Type')
plt.draw()
```
